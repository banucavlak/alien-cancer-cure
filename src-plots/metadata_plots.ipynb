```python
import IPython

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
```

### **1 Reading Metadata File**

```python
df_meta = pd.read_csv('../data/metadata/str_metadata.tsv', sep='\t')
```

```python
df_meta.head()
```

```python
import seaborn as sns

fig, ax = plt.subplots()

sns.set_theme(style="whitegrid")

g = sns.histplot(
    data=df_meta, x="alleged_abduction", hue="disease", multiple='dodge', discrete=True, ax=ax, shrink=0.8
)
#plt.legend(fontsize='large')


#xtick_labels = ['non abducted', 'abducted']
#ax.set_xticklabels(xtick_labels, fontsize=10);
ax.set_xlabel('Alien Abduction Status', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
#ax.set_title('Distribution of Diseases across Alien Abduction', fontsize=16);
#sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.setp(ax.get_legend().get_texts(), fontsize='8') # for legend text


plt.savefig("../plots/metadata/metadata_disease" + ".pdf", dpi=600, format='pdf', transparent=False, bbox_inches='tight')
plt.show()
```
